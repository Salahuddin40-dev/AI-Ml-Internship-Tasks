```python
#import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load dataset
df = sns.load_dataset("iris")
#print shape
print(df.shape)
#print column names
print(df.columns.tolist())
# print first five rows
df.head
```

```python
# display summary information
df.info()
```

```python
#display descriptive statistics
df.describe()
```

```python
#scatterplot showing reltions between features
sns.scatterplot(data = df, x="sepal_length",y = "sepal_width" , hue = "species")
plt.title("relationship between sepal length and sepal width")
plt.xlabel("sepal_length (cm)")
plt.ylabel("sepal_width (cm)")
plt.legend(title = "species")
plt.show()

sns.scatterplot(data = df, x="petal_length",y = "petal_width" , hue = "species")
plt.title("relationship between petal length and petal width")
plt.xlabel("petal_length (cm)")
plt.ylabel("petal_width (cm)")
plt.legend(title = "species")
plt.show()
```

```python
#histogram showing value distribution
df.hist(figsize = (10,8), bins = 15)
plt.suptitle("value distribtion" , fontsize = 16)
plt.tight_layout()
plt.show()
```

```python
#box plot to identify outliers
plt.figure(figsize = (12,6))
for i , column in enumerate(df.select_dtypes(include = 'number').columns):
    plt.subplot(1,4,i+1)
    sns.boxplot(y = df[column])
    plt.title(column)
    plt.tight_layout()
    plt.show()
```
